```python
from PIL import ImageGrab
import cv2
import numpy as np
import time

WIDTH = 640
HEIGHT = 480


def capture_screen():
    screen = ImageGrab.grab(bbox=(0, 0, WIDTH, HEIGHT))
    screen = np.array(screen)
    screen = cv2.cvtColor(screen, cv2.COLOR_BGR2RGB)
    return screen


def main():
    while True:
        frame_time = time.time()
        cv2.imshow("window", capture_screen())
        frame_time = time.time() - frame_time

        print("FPS: ", 1 / frame_time)

        if cv2.waitKey(25) & 0xFF == ord("q"):
            cv2.destroyAllWindows()
            break
```

```python
main()
```

```python
import numpy as np
import cv2
from mss import mss

bounding_box = {"top": 100, "left": 0, "width": 640, "height": 480}

sct = mss()

while True:
    frame_time = time.time()
    sct_img = sct.grab(bounding_box)
    frame_time = time.time() - frame_time
    cv2.imshow("screen", np.array(sct_img))

    print("FPS: ", 1 / frame_time)

    if (cv2.waitKey(1) & 0xFF) == ord("q"):
        cv2.destroyAllWindows()
        break
```
